##Problema 1.
¿Qué es o que significa la cantidad $||\theta^*||^2\over{\delta^2}$?

En mi opinión, se relaciona con la siguiente cota, dada una iteración $k$ se tiene que: 
$$(\theta^*)^T\theta^k \geq k\delta$$
Como sabemos que el algoritmo encuentra el $\theta$ correcto si existe, entonces para alguna iteración $K$ se debe tener que $$\theta^* = \theta^K$$
Así tenemos que:
$${||\theta^*||^2\over{\delta}} \geq K $$
Una cota para K, entonces el valor 
$||\theta^*||^2\over{\delta^2}$ es proporcional al mínimo número de iteraciones que debo hacer para obtener una solución.

###(a)
Como $x(t)$ no fue clasificado correctamente por $w(t)$ ocurre que $y(t)$ y $w^T(t)x(t)$ tienen signos opuestos, así si $y(t) = 1$, entonces $w^T(t)x(t) < 0$. Por otro lado, si $y(t) = -1$, entonces $w^T(t)x(t) > 0$. En ambos casos, tenemos que  su producto debe ser menor a $0$.

###(b)


Tenemos que:
\begin{equation}
\begin{split}
y(t)w^T(t+1)x(t)
&= y(t)(w(t)+y(t)x(t))^Tx(t)  \\
&=  y(t)((w^T(t)x(t))+y(t)x^T(t)x(t))  \\
&=  y(t)w^T(t)x(t)+y^2(t)x^T(t)x(t)  \\
&> y(t)w^T(t)x(t)\\
\end{split}
\end{equation}
Donde $y^2(t)x^T(t)x(t) > 0$ pues $y^2(t) = 1$ y como $w^T(x)x(t) \neq 0$ entonces $x(t) \neq 0$ y $||x(t)||^2 = x^T(t)x(t) > 0$.

###(c) 

Si $x(t)$ no fue clasificado correctamente entonces $y(t)w^T(t)x(t) < 0$. Nuestro objetivo es que $x(t)$ sea clasificado correctamente, es decir, que el producto $y(t)w^T(t)x(t)$ sea positivo, luego al tomar $w(t+1)$ tenemos que $y(t)w^T(t+1)x(t) > y(t)w^T(t)x(t)$ esto significa que $y(t)w^T(t+1)x(t)$ está mas cerca de $0$ que $y(t)w^T(t)x(t)$  si el producto resultó ser negativo, o es positivo. En ambos casos vemos que $w(t+1)$ resulta ser una mejoría de $w(t)$ para la clasificación de $x(t)$.

```python
#Simulación 1000 monedas
monedas = 1000
lanzamientos = 10
mu = 0.5
np.random.seed(13) #Seed
coins = np.zeros((monedas,lanzamientos)) #Array de resultados
for i in range(1000):
  coins[i] = random.binomial(1,mu,lanzamientos)
v1 =  np.sum(coins[0])/lanzamientos
i2 =  ran.randrange(0,monedas)
vr =  (np.sum(coins[i2]))/lanzamientos
vm = lanzamientos
for i in range(monedas): #Buscamos el indice de menor monedas
  count = np.sum(coins[i])
  if count < vm:
    vm = count
vm /= lanzamientos
print("V1 = ",v1,"Vr = ",vr,"Vmin = ",vm)
```

###b)

```python
#Repetimos el experimento
np.random.seed(20) #Seed
iteraciones = 100000
monedas = 1000
lanzamientos = 10
mu = 0.5
v1 = np.zeros(iteraciones)
vr = np.zeros(iteraciones)
vm = np.zeros(iteraciones)
for k in range(iteraciones):
  #Simulación 1000 monedas
  coins = np.ones((monedas, lanzamientos), dtype=np.int64) *  random.binomial(1, .5, (monedas, lanzamientos)) #Array de resultados
  v1[k] =  np.sum(coins[0])/lanzamientos
  i2 =  ran.randrange(0,monedas)
  vr[k] =  (np.sum(coins[i2]))/lanzamientos
  vm[k] = lanzamientos
  for i in range(monedas): #Buscamos el indice de menor monedas
    count = np.sum(coins[i])
    if count < vm[k]:
      vm[k] = count
  vm[k] /= lanzamientos
```

###c)

```python
def prob(eps, mu, v):
  p = len(v)
  less = 0
  for i in range(p):
    if abs(mu-v[i]) > eps:
      less = less + 1
  return less/(p+0.0)
def hoeff(eps, n):
  return 2*np.exp(-2*(eps**2)*n)
siz = 1000
max_e = 0.9
epsval = np.zeros(siz)
probv1 = np.zeros(siz)
probvm = np.zeros(siz)
probvr = np.zeros(siz)
hoeffding = np.zeros(siz)
z = 0
for i in range(siz):
  epsval[i] = z
  probv1[i] = prob(epsval[i], mu, v1)
  probvm[i] = prob(epsval[i], mu, vm)
  probvr[i] = prob(epsval[i], mu, vr)
  hoeffding[i] = hoeff(epsval[i],lanzamientos)
  z += max_e/siz
```

```python
plt.figure(figsize=(10, 5))     
#plt.scatter(x,y)
plt.plot(probv1, epsval, 'b-',label='v_1')
plt.plot(probvr, epsval, 'r-',label='v_ran')
plt.plot(probvm, epsval, 'g-',label='v_min')
plt.plot(hoeffding, epsval, 'y-',label='Hoeffding')
plt.title("Estimaciones para la probabilidad")
legend = plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.xlabel("Epsilon")
plt.ylabel("Probabilidad")
plt.show()
```
